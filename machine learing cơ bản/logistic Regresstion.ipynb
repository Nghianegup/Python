## Leaky relu

```python
x = plt.linspace(-10,10,100)
a = 0.8
b = -4
y = a*x + b
def leakyrelu(x):
    j = len(y)
    for j in range(j):
        if y[j] < 0 : 
            y[j] = 0.01*y[j]
        else :
            y[j] = y[j]   
    return y    
    print(y)    
y = leakyrelu(x)  
print(y)
plt.grid()
plt.plot(x,y,'r')    
```

## Binary step

```python
a = 0.8
b = -4
x = plt.linspace(-10,10,100)
y = a*x +b
def Binarystep():
    j = len(y)
    for j in range(j):
        if y[j] < 0:
            y[j] = 0
        else:
            y[j]= 1 
    return y
    print(y)
h = Binarystep()
plt.plot(x,h,'r')
    
```

## Inverse square root linear unit 

```python
a = 0.8
b =- 4
c = 2
x = plt.linspace(-10,10,100)
y = a*x + b
def isrlu():
    j = len(y)
    for j in range(j):
        if y[j] < 0:
            y[j] = y[j]/np.sqrt(1+c*(y[j]**2))
        else:
            y[j] = y[j]
    print(y)
    return y
i = isrlu()
plt.plot(x,i,'r')
    
```

## Elu 

```python
a = 1
b = -20
c = 3
x = plt.linspace(-10,10,100)
y = a*x +b
def ELU():
    j = len(y)
    for j in range(j):
        if y[j] <= 0:
            y[j] = c*(np.exp(y[j])-1)
        else:
            y[j]= y[j]
    return y
h = ELU()
plt.plot(x,h,'r')
```
