## Bài tập biến đổi hình 

```python
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
from numpy import random

```

#### Quay hình 90 độ theo chiều kim đồng hồ 

```python
X = img.imread('real.jpg')
print(X)
print(np.shape(X))
plt.subplot(231)
plt.imshow(X)
plt.show()
```

#### Quay hình 90 độ theo chiều kim đồng hồ 

```python
A = np.rot90(X,-2)
plt.imshow(A)
```

#### Quay hình 90 độ ngược chiều kim đồng hồ 

```python
A = np.rot90(X,1)
plt.imshow(A)
```

#### lật hình dọc 

```python
plt.imshow(X)
plt.imshow(X[::-1])
print(X[2,3,1])
print(X[2,3,2])
print(X[2,3,0])
```

#### lật hình ngang 

```python
plt.imshow(X[:,::-1])
```
